# Data Cleaning

#### Dealing with null values

```python
less_than_4_percent_missing = missing_proportion[missing_proportion < 4].index
more_than_5_percent_missing = missing_proportion[missing_proportion > 5].index

# Remove rows with missing values in columns with less than 1% missing values
df = df.dropna(subset=less_than_4_percent_missing)

# Imputation methods
# Columns with more than 5% missing values
for column in more_than_5_percent_missing:
    if df[column].dtype == 'object':
        # Impute missing values with the mode (most frequent value) for categorical columns
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        # Impute missing values with the median for numerical columns
        df[column].fillna(df[column].median(), inplace=True)
```

```python
df.isnull().sum()
```

#### Choosing only one main genre foe each column

```python
df['primary_genre'] = df['genres'].str.split('|').str[0] # I will choose the first genre as he first genre listed is considered the most relevant in mose cases
df['primary_genre']
```

#### Adding **Net Revenue** column

```python
df['Net revenue']=df['gross']-df['budget']
df['Net revenue']
```

#### Removing outliers

```python
def identify_outliers(df, column_names):
    for column_name in column_names:
        while True:
            Q1 = df[column_name].quantile(0.25)
            Q3 = df[column_name].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            initial_shape = df.shape[0]
            df = df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)]
            final_shape = df.shape[0]
            if initial_shape == final_shape:
                break
    return df

df = identify_outliers(df, ['duration'])
```

### How the rating of movies changed over years?

```python
df.groupby('title_year')['imdb_score'].mean().plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Average Rating')
```

The rating of movies were considerably constant till **1960**, then there was an increase in the ratings between **1960** and **1970**, after that the ratings were in gradual decrease until **2020**. This suggest that the quality and types of new movies are not as good as old movies

### What is the average rating of each movie genre?

```python
df.groupby('primary_genre')['imdb_score'].mean().plot(kind='bar',figsize=(12,6))
plt.xlabel('Genre')
plt.ylabel('Average Rating')
```

The **Film-Noir** genre stands out with the highest average rating

### Revenue generated by each director

```python
df.groupby('director_name')['Net revenue'].sum().sort_values(ascending=False).head(10).plot(kind='bar',figsize=(12,6))
plt.xlabel('Director')
plt.ylabel('Revenue')
```

### What are the top 10 countries in movie production?
